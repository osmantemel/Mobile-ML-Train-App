```python
import pandas as pd

def fill_all_missing_values(df, fill_value=None):

    # İlk olarak, fill_value ile doldur
    if fill_value is not None:
        df = df.fillna(fill_value)
    
    # Daha sonra forward fill ve backward fill uygula
    df = df.fillna(method='ffill')
    df = df.fillna(method='bfill')
    
    # Hala boş kalan hücreler varsa, bunları fill_value ile doldur
    if fill_value is not None:
        df = df.fillna(fill_value)
    
    return df


df = fill_all_missing_values(df, fill_value=0)
print("Boş verileri doldurulmuş DataFrame:")
print(df)
```

```python
import pandas as pd
import random

# Veri seti için boş bir DataFrame oluşturma
df = pd.DataFrame(columns=['ID', 'Name', 'Age', 'Gender', 'Score'])

# Veri setindeki satır sayısı
num_rows = 100

# Veri setini döngüyle oluşturma
for i in range(num_rows):
    # ID (1'den başlayarak sıralı bir şekilde artan değerler)
    ID = i + 1
    
    # İsim (rastgele seçilen bir isim)
    names = ['Alice', 'Bob', 'Charlie', 'David', 'Emma', 'Frank', 'Grace']
    name = random.choice(names)
    
    # Yaş (18 ile 80 arasında rastgele bir yaş)
    age = random.randint(18, 80)
    
    # Cinsiyet (rastgele seçilen bir cinsiyet)
    genders = ['Male', 'Female']
    gender = random.choice(genders)
    
    # Skor (0 ile 100 arasında rastgele bir sayı)
    score = random.randint(0, 100)
    
    # Yeni satırı veri çerçevesine ekleme
    df.loc[len(df)] = [ID, name, age, gender, score]

# Oluşturulan veri setini gösterme
print(df.head())
```
